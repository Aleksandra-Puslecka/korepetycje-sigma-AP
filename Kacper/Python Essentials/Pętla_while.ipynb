## Pętla `while`?

Pętla `while` jest jednym z podstawowych mechanizmów iteracji w Pythonie. Działa na podstawie warunku logicznego: dopóki warunek jest `True`, kod wewnątrz pętli będzie się wykonywał.

### Składnia pętli `while`

```python
while warunek:
    # kod do wykonania
```

### Kluczowe elementy:
- **Warunek**: Jest to wyrażenie logiczne, które decyduje, czy pętla ma się kontynuować.
- **Kod do wykonania**: Jest to blok kodu, który zostanie wykonany w każdej iteracji.

---


````python

# Przykład podstawowy
licznik = 0

while licznik < 5:
    print(f"Licznik wynosi: {licznik}")
    licznik += 1  # Zwiększamy licznik o 1
````


## Kluczowe aspekty pętli `while`

### 1. Nieskończona pętla
Jeżeli warunek nigdy nie stanie się `False`, pętla będzie wykonywać się bez końca.

**Przykład:**
```python
while True:
    print("To się nigdy nie skończy!")
```

Aby uniknąć nieskończonych pętli, upewnij się, że istnieje mechanizm, który zmienia warunek na `False`. Najczęściej jest to zmiana wartości zmiennej używanej w warunku.

### 2. Instrukcje `break` i `continue`
- **`break`**: Przerywa działanie pętli.
- **`continue`**: Przechodzi do następnej iteracji, pomijając kod poniżej.

**Przykład `break`:**
```python
licznik = 0
while licznik < 10:
    if licznik == 5:
        break
    print(licznik)
    licznik += 1
```

**Przykład `continue`:**
```python
licznik = 0
while licznik < 5:
    licznik += 1
    if licznik == 3:
        continue
    print(licznik)
```

### 3. Zagnieżdżone pętle `while`
Możemy używać pętli `while` wewnątrz innych pętli `while`.

**Przykład:**
```python
x = 0
while x < 3:
    y = 0
    while y < 2:
        print(f"x = {x}, y = {y}")
        y += 1
    x += 1
```

## Ćwiczenia

1. **Prosta pętla**
Napisz pętlę `while`, która wypisze liczby od 10 do 1 w odwrotnej kolejności.

2. **Zgadnij liczbę**
Napisz program, który losuje liczbę z zakresu od 1 do 10. Użytkownik ma zgadnąć tę liczbę, a program powinien informować, czy odpowiedź jest poprawna. Wykorzystaj pętlę `while`.

3. **Nieskończona pętla z przerwaniem**
Stwórz nieskończoną pętlę `while`, która prosi użytkownika o podanie hasła. Przerwij pętlę, gdy użytkownik poda poprawne hasło.

4. **Zagnieżdżona pętla**
Utwórz zagnieżdżoną pętlę `while`, która generuje tabliczkę mnożenia dla liczb od 1 do 5.

---

Rozwiązania znajdziesz w kolejnych komórkach poniżej.


````python

# 1. Prosta pętla
licznik = 10
while licznik > 0:
    print(licznik)
    licznik -= 1

# 2. Zgadnij liczbę
import random
wylosowana = random.randint(1, 10)
zgadnieto = False

while not zgadnieto:
    odpowiedz = int(input("Zgadnij liczbę (od 1 do 10): "))
    if odpowiedz == wylosowana:
        print("Brawo! Zgadłeś.")
        zgadnieto = True
    else:
        print("Spróbuj jeszcze raz.")

# 3. Nieskończona pętla z przerwaniem
poprawne_haslo = "tajne"
while True:
    haslo = input("Podaj hasło: ")
    if haslo == poprawne_haslo:
        print("Hasło poprawne!")
        break
    else:
        print("Niepoprawne hasło. Spróbuj ponownie.")

# 4. Zagnieżdżona pętla (tabliczka mnożenia)
x = 1
while x <= 5:
    y = 1
    while y <= 5:
        print(f"{x} x {y} = {x * y}")
        y += 1
    x += 1
````
